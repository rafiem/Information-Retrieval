```python
import pandas as pd
import numpy as np
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import LancasterStemmer
```

```python
df = pd.read_excel('databook.xlsx')
df.columns=['Wikipedia article ID',
           'Freebase ID',
           'Book title',
           'Author',
           'Date',
           'Genres ID',
            'Genres',
           'Summary']
df
```

```python
to_drop = ['Wikipedia article ID','Freebase ID', 'Genres ID','Date'] #drop column
df.drop(to_drop, inplace = True, axis = 1)
df.head()
```

```python
df=df.dropna() #drop data NaN
df
```

```python
def preprocessing(row):
    filtered_sentence = []
    stem_sentence =[]
    row = row.lower()
    words = word_tokenize(row)
    stopWords = set(stopwords.words('english'))
    lancaster=LancasterStemmer()
    for w in words:
        if w not in stopWords: #stopword
            filtered_sentence.append(w)
            stem_sentence.append(lancaster.stem(w)+" ")
    return "".join(stem_sentence)
```

```python
outSummary=[]
for row in df['Summary']:
    outSummary.append(preprocessing(row))
```
